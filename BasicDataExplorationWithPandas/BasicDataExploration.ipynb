Temos dois tipos de entidades centrais em Pandas: 
* DataFrames

O DrataFrame é como uma tabela que contem um conjunto entradas (dados), que são formatados em linhas e colunas. 
No exemplo abaixo temos um dataframe. As linhas são identificadas de forma que a primeira é sempre 0. Nas colunas ou headers, podemos especificar a natureza do dado fornecido.

* Series
  
A série é como uma lista de dados. Os dados ficam sequenciados, por ordem de "chegada", como estamos acostumados.

```python
# What is the average lot size (rounded to nearest integer)?
avg_area_size = round(melbourne_data['Landsize'].mean())
print("A média das áreas das casas é:  ",avg_area_size , "m"+ '\u00B2')

# As of today, how old is the newest home (current year - the date in which it was built)
newest_home_age = 2025 - melbourne_data.YearBuilt.max()
print("A casa construida ha menos tempo possui ", newest_home_age, " anos.")
```

```python
# Index in Pandas = first line of csv
melbourne_data.iloc[0]
```

```python
# Index in Pandas = first line of csv
# On its own, the : operator, which also comes from native Python, means "everything"
melbourne_data.iloc[:, 0]
```

```python
#  When combined with other selectors, however, it can be used to indicate a range of values. 
# For example, to select the first column from just the first, second, and third row, we would do:
melbourne_data.iloc[:3, 0]
```

```python
# Or, to select just the second and third entries
melbourne_data.iloc[1:3, 0]
```
